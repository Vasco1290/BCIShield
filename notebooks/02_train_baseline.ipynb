# 02 - Train Baseline

Train the baseline EEGNet mode on clean data.

```python
import sys
sys.path.append('..')
import torch
from src.models.eegnet import EEGNet
from src.data.dataset import get_dataloaders
from experiments.run_experiment import train_base_model, evaluate_model

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
train_loader, test_loader = get_dataloaders('../data/raw/', 1, batch_size=32)

model = EEGNet().to(device)
print("Training...")
model = train_base_model(model, train_loader, epochs=30, lr=0.001, device=device)

acc = evaluate_model(model, test_loader, device=device)
print(f"Clean accuracy: {acc:.2f}%")
```
