# 01 - Data Exploration

Visually inspect the EEG data and plot the evoked potentials.

```python
import sys
sys.path.append('..')
import mne
import torch
import matplotlib.pyplot as plt
from src.data.dataset import BCIDataset

# Ensure dataset is downloaded to data/raw/
dataset = BCIDataset(data_dir='../data/raw', subject_id=1, is_train=True)
print(f"Loaded {len(dataset)} trials")
```

```python
# Plot a single trial (channel 0)
x, y = dataset[0]
plt.figure(figsize=(10, 4))
plt.plot(x[0, 0, :].numpy())
plt.title(f"EEG Trial for subject 1, clean, label {y}")
plt.xlabel("Time samples")
plt.ylabel("Normalized Amplitude")
plt.show()
```
